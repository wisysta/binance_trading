```python
#라벨인덱싱 : 인덱스 사용
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5, 4), 
                   index=pd.date_range('20210101', periods=5) , 
                   columns=["a","b","c","d"])
print("*df:\n",df,"\n")

df1 = df.loc['20210102':'20210104'] #(1)
print("*df1\n", df1, "\n")
df2 = df.loc[:'20210104'] #(2)
print("*df2\n", df2, "\n")
```

```python
#라벨인덱싱 :  컬럼과 인덱스 사용

df1 = df.loc['20210102':'20210104', 'b':'c'] #(1)
print("*df1\n", df1, "\n")
df2 = df.loc[:'20210104', :'c'] #(2)
print("*df2\n", df2, "\n")
df3 = df.loc[:'20210104', ['a','c']] #(3)
print("*df3\n", df3, "\n")
```

```python
#라벨인덱싱 :  조건문 사용

print(df,"\n")

df1 = df.loc['2021-01-03'] > 0.1 #(1)
print("*df1:\n",df1,"\n")

df2 = df.loc[:, df1] #(2)
print("*df2:\n",df2,"\n")
```

```python
#위치인덱싱

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5, 4), 
                   index=pd.date_range('20210101', periods=5) , 
                   columns=["a","b","c","d"])
print("*df:\n",df,"\n")

df1 = df.iloc[1:3] #(1)
print("*df1\n", df1, "\n")
df2 = df.iloc[:3] #(2)
print("*df2\n", df2, "\n")
```
